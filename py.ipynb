```python
def rotate_array(arr, k):
    n = len(arr)
    if n == 0:  # Handle empty array case
        return []

    # Ensure k is within the bounds of the array length
    # A rotation by n positions brings the array back to its original state
    k = k % n

    # Perform the rotation using slicing
    # The last k elements move to the front, and the remaining elements follow
    rotated_arr = arr[-k:] + arr[:-k]
    return rotated_arr

# Example Usage:
my_array = [1, 2, 3, 4, 5, 6, 7]
k_value = 3

rotated_result = rotate_array(my_array, k_value)
print(f"Original Array: {my_array}")
print(f"Rotated Array by {k_value} positions: {rotated_result}")

# Another example
my_array_2 = ['a', 'b', 'c', 'd']
k_value_2 = 1
rotated_result_2 = rotate_array(my_array_2, k_value_2)
print(f"\nOriginal Array: {my_array_2}")
print(f"Rotated Array by {k_value_2} positions: {rotated_result_2}")

# Example with k > len(arr)
my_array_3 = [10, 20, 30]
k_value_3 = 4
rotated_result_3 = rotate_array(my_array_3, k_value_3)
print(f"\nOriginal Array: {my_array_3}")
print(f"Rotated Array by {k_value_3} positions: {rotated_result_3}")
```

```python
def gcd(a, b):
    # Ensure numbers are positive
    if a <= 0 or b <= 0:
        print("Please enter positive integers for GCD calculation.")
        return None

    while b:
        a, b = b, a % b
    return a

num1 = int(input("Enter the first positive number: "))
num2 = int(input("Enter the second positive number: "))

result = gcd(num1, num2)

if result is not None:
    print(f"The GCD/HCF of {num1} and {num2} is: {result}")
```

```python
def gcd(a, b):
    # Helper function to find the Greatest Common Divisor (GCD)
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    # Ensure numbers are positive
    if a <= 0 or b <= 0:
        print("Please enter positive integers for LCM calculation.")
        return None

    # LCM(a, b) = |a*b| / GCD(a, b)
    # Since we are dealing with positive numbers, abs() is not strictly needed
    return (a * b) // gcd(a, b)

num1 = int(input("Enter the first positive number: "))
num2 = int(input("Enter the second positive number: "))

result = lcm(num1, num2)

if result is not None:
    print(f"The LCM of {num1} and {num2} is: {result}")
```

```python
print("--- Method 1: Line-by-line Input ---")
while True:
    try:
        num_elements = int(input("Enter the number of elements for the first list: "))
        if num_elements <= 0:
            print("Please enter a positive integer.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter an integer.")

list_line_by_line = []
print(f"Enter {num_elements} elements, one by one:")
for i in range(num_elements):
    while True:
        try:
            element = int(input(f"Element {i+1}: "))
            list_line_by_line.append(element)
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

print(f"List built element by element: {list_line_by_line}")

print("\n--- Method 2: Space-separated Input ---")
while True:
    space_separated_input = input("Enter elements for the second list, separated by spaces: ")
    if not space_separated_input.strip(): # Check if input is empty or just whitespace
        print("Input cannot be empty. Please enter some elements.")
        continue

    # Split the input string by spaces
    elements_as_strings = space_separated_input.split()

    # Convert elements to integers, handling potential errors
    list_space_separated = []
    conversion_error = False
    for s in elements_as_strings:
        try:
            list_space_separated.append(int(s))
        except ValueError:
            print(f"Invalid input: '{s}' is not a valid integer. Please enter only integers separated by spaces.")
            conversion_error = True
            break

    if not conversion_error:
        break

print(f"List built from space-separated input: {list_space_separated}")
```

```python
def linear_search(data_list, target_element):
    """
    Performs a linear search to find the target_element in data_list.
    Returns the index of the target_element if found, otherwise returns -1.
    """
    for i in range(len(data_list)):
        if data_list[i] == target_element:
            return i
    return -1

# Create a sample list
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

# Example 1: Target element is present in the list
target1 = 30
index1 = linear_search(my_list, target1)
if index1 != -1:
    print(f"'{target1}' found at index: {index1}")
else:
    print(f"'{target1}' not found in the list.")

# Example 2: Target element is not present in the list
target2 = 60
index2 = linear_search(my_list, target2)
if index2 != -1:
    print(f"'{target2}' found at index: {index2}")
else:
    print(f"'{target2}' not found in the list.")
```

```python
print("--- Demonstrating Variable Swapping ---")

# 1. Define two variables and assign initial integer values
var1 = 10
var2 = 20

# 2. Print the initial values of var1 and var2
print(f"Initial values: var1 = {var1}, var2 = {var2}")

# 3. Demonstrate the first method of swapping using a temporary variable
print("\nMethod 1: Swapping using a temporary variable")
temp = var1  # a. Assign the value of var1 to a temporary variable
var1 = var2  # b. Assign the value of var2 to var1
var2 = temp  # c. Assign the value of temp to var2

# 4. Print the values of var1 and var2 after the first swap
print(f"After first swap: var1 = {var1}, var2 = {var2}")

# 5. Reset var1 and var2 to their original initial values
var1 = 10
var2 = 20

# 6. Print the reset values of var1 and var2
print(f"\nValues reset: var1 = {var1}, var2 = {var2}")

# 7. Demonstrate the second method of swapping using tuple packing and unpacking
print("\nMethod 2: Swapping using tuple packing/unpacking (Pythonic way)")
var1, var2 = var2, var1 # a. Assign var2, var1 to var1, var2

# 8. Print the values of var1 and var2 after the second swap
print(f"After second swap: var1 = {var1}, var2 = {var2}")
```

```python
print("--- Demonstrating Reversal ---")

# 1. Initialize a sample list and a sample string
my_list = [1, 2, 3, 4, 5]
my_string = "Python"

# 2. Print the original list and string
print(f"Original list: {my_list}")
print(f"Original string: '{my_string}'")

# 3. Demonstrate reversing the list using the reverse() method
# This method modifies the list in-place and does not return a new list.
print("\nMethod 1: Reversing list using list.reverse() (in-place modification)")
my_list.reverse()
print(f"List after reverse() method: {my_list}")
print("Note: The original 'my_list' object has been modified directly.")

# 4. Reset the list to its original state for the next demonstration
my_list = [1, 2, 3, 4, 5]
print(f"\nList reset to original: {my_list}")

# 4. Demonstrate reversing the list using slicing (creates a new list)
print("\nMethod 2: Reversing list using slicing [::-1] (creates a new list)")
reversed_list_slicing = my_list[::-1]
print(f"Original list (unmodified by slicing): {my_list}")
print(f"New reversed list using slicing: {reversed_list_slicing}")
print("Note: Slicing creates a new list, leaving the original list unchanged.")

# 5. Demonstrate reversing the string using slicing (creates a new string)
print("\nMethod 3: Reversing string using slicing [::-1] (creates a new string)")
reversed_string_slicing = my_string[::-1]
print(f"Original string (unmodified by slicing): '{my_string}'")
print(f"New reversed string using slicing: '{reversed_string_slicing}'")
print("Note: Strings are immutable in Python, so slicing always creates a new string.")
```

```python
print("\n--- Demonstrating `else` with `for` loop ---")

# Example for loop with else block (item not found)
my_list = [10, 20, 30, 40, 50]
target_for = 60
found_for = False

print(f"Searching for {target_for} in {my_list}")
for item in my_list:
    if item == target_for:
        print(f"Found {target_for} in the list.")
        found_for = True
        break
else:
    print(f"{target_for} not found in the list (else block of for loop executed).")

# Example for loop with else block (item found)
target_for_found = 30
print(f"\nSearching for {target_for_found} in {my_list}")
for item in my_list:
    if item == target_for_found:
        print(f"Found {target_for_found} in the list.")
        break
else:
    print(f"{target_for_found} not found in the list (else block of for loop executed).")


print("\n--- Demonstrating `else` with `while` loop ---")

# Example while loop with else block (normal completion)
count_while = 0
limit_while = 3

print(f"Counting up to {limit_while} with a while loop:")
while count_while < limit_while:
    print(f"Current count: {count_while}")
    count_while += 1
else:
    print("While loop completed normally (else block of while loop executed).")

# Example while loop with else block (terminated by break)
count_break = 0
break_at = 1

print(f"\nCounting, but breaking at {break_at}:")
while count_break < limit_while:
    print(f"Current count: {count_break}")
    if count_break == break_at:
        print("Breaking out of the loop.")
        break
    count_break += 1
else:
    print("While loop completed normally (else block of while loop executed).")
```

```python
print("--- Demonstrating String Operations and Methods ---")

# 1. Initialize a sample string variable with some text.
my_string = "Python Programming is Fun!"

# 2. Print the original string.
print(f"\nOriginal string: '{my_string}'")

# 3. Demonstrate basic string operations:
# a. Find the length of the string using len().
string_length = len(my_string)
print(f"\nLength of the string: {string_length}")

# b. Convert the string to uppercase using .upper().
uppercase_string = my_string.upper()
print(f"Uppercase string: '{uppercase_string}'")

# c. Convert the string to lowercase using .lower().
lowercase_string = my_string.lower()
print(f"Lowercase string: '{lowercase_string}'")

# d. Capitalize the first letter of the string using .capitalize().
capitalized_string = my_string.capitalize()
print(f"Capitalized string: '{capitalized_string}'")

# e. Replace a substring within the string using .replace().
replaced_string = my_string.replace("Fun", "Awesome")
print(f"String after replacement: '{replaced_string}'")

# f. Check if a substring is present using the 'in' operator.
substring_present = "Python" in my_string
print(f"Is 'Python' present in the string? {substring_present}")

substring_not_present = "Java" in my_string
print(f"Is 'Java' present in the string? {substring_not_present}")

# g. Find the starting index of a substring using .find().
find_index = my_string.find("Programming")
print(f"Starting index of 'Programming': {find_index}")

find_non_existent = my_string.find("hello")
print(f"Starting index of 'hello' (not found): {find_non_existent}")

# h. Split the string into a list of words using .split().
split_string = my_string.split()
print(f"String split into words: {split_string}")

# i. Join a list of strings into a single string using .join().
words_to_join = ["Hello", "World", "from", "Python"]
joined_string = "-".join(words_to_join)
print(f"Joined string with '-': '{joined_string}'")

joined_string_space = " ".join(split_string)
print(f"Re-joined original split string with space: '{joined_string_space}'")
```

```python
print("--- Demonstrating Dictionary Operations and Methods ---")

# 1. Initialize a sample dictionary with a few key-value pairs.
my_dict = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

# 2. Print the entire dictionary.
print(f"\nOriginal dictionary: {my_dict}")

# 3. Demonstrate accessing a value by its key.
print(f"Name: {my_dict['name']}")
print(f"Age: {my_dict['age']}")

# 4. Demonstrate adding a new key-value pair to the dictionary.
my_dict['occupation'] = 'Engineer'
print(f"\nDictionary after adding 'occupation': {my_dict}")

# 5. Demonstrate modifying the value associated with an existing key.
my_dict['age'] = 31
print(f"Dictionary after modifying 'age': {my_dict}")

# 6. Demonstrate using the .keys() method to get all keys and print them.
keys_list = list(my_dict.keys())
print(f"\nAll keys: {keys_list}")

# 7. Demonstrate using the .values() method to get all values and print them.
values_list = list(my_dict.values())
print(f"All values: {values_list}")

# 8. Demonstrate using the .items() method to get all key-value pairs and print them.
items_list = list(my_dict.items())
print(f"All key-value pairs: {items_list}")

# 9. Demonstrate checking if a key exists in the dictionary using the 'in' operator.
key_exists_name = 'name' in my_dict
key_exists_country = 'country' in my_dict
print(f"\nDoes 'name' exist in the dictionary? {key_exists_name}")
print(f"Does 'country' exist in the dictionary? {key_exists_country}")
```

```python
print("--- Demonstrating Sorting Operations ---")

# 1. Initialize a sample list of numbers
my_numbers = [5, 2, 8, 1, 9, 3]

# 2. Print the original list to show its unsorted state.
print(f"\nOriginal list: {my_numbers}")

# 3. Demonstrate sorting the list in-place using the .sort() method.
print("\nSorting list in-place using .sort():")
my_numbers.sort()
# Print the list again after this operation.
print(f"List after .sort(): {my_numbers}")
print("Note: The original 'my_numbers' object has been modified directly.")

# 4. Reset the list to its original unsorted state.
my_numbers = [5, 2, 8, 1, 9, 3]
print(f"\nList reset to original: {my_numbers}")

# 5. Demonstrate sorting the list using the sorted() built-in function.
print("\nSorting list using sorted() function (creates a new list):")
new_sorted_list = sorted(my_numbers)
# Print both the original (unmodified) list and the newly created sorted list.
print(f"Original list (unmodified by sorted()): {my_numbers}")
print(f"New sorted list created by sorted(): {new_sorted_list}")
print("Note: The sorted() function returns a new list, leaving the original list unchanged.")
```

```python
print("--- Demonstrating 2-Dimensional Lists ---")

# 1. Initialize a 2-dimensional list (list of lists) with some sample data.
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# 2. Print the entire 2-dimensional list to display its structure.
print(f"\nOriginal 2D List (Matrix): {matrix}")

# 3. Demonstrate how to access a specific element using its row and column indices.
#    Accessing element at row 1, column 2 (0-indexed: matrix[1][2] which is 6)
row_index = 1
col_index = 2
specific_element = matrix[row_index][col_index]
print(f"\nAccessing element at row {row_index}, column {col_index}: {specific_element}")

# 4. Demonstrate how to access an entire row of the 2-dimensional list.
#    Accessing row at index 0 (0-indexed: matrix[0] which is [1, 2, 3])
row_to_access = 0
entire_row = matrix[row_to_access]
print(f"Accessing row at index {row_to_access}: {entire_row}")

# 5. Iterate through the 2-dimensional list using nested loops to print each element.
#    Ensure to print them in a way that visualizes the 2D structure.
print("\nIterating through 2D list (visualizing structure):")
for row in matrix:
    for element in row:
        print(element, end=" ") # Print elements of a row on the same line
    print() # Move to the next line after each row

# 6. Iterate through the 2-dimensional list using nested loops to print each element
#    along with its row and column indices.
print("\nIterating through 2D list with indices:")
for i in range(len(matrix)): # i represents the row index
    for j in range(len(matrix[i])): # j represents the column index
        print(f"Element at ({i}, {j}): {matrix[i][j]}")
```
